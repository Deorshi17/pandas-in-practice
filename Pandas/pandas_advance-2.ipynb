In Pandas, you can access data in 2 ways:

1️⃣ Implicit Indexing (by position)

👉 Uses index numbers (0, 1, 2, …)

👉 Like arrays in Python

🔹 Functions used:
- iloc[] → integer location

2️⃣ Explicit Indexing (by labels/names)

👉 Uses row names and column names

🔹 Functions used:
- loc[] → label-based



**Implicit indexing** also known as Position-based indexing (iloc)

```python
df.iloc[0:2] #start from 0 and go till 1
# selects first two rows (index 0 and 1) using position-based indexing
```

```python
df.loc[0:2] #loc is used for named indexes, iloc is used for integer indexes
```

index ka koi bhi naam rkh do jaise ab, bc, cd to usko find krne ka method hai df.loc['ab':'cd'] 
(kyunki loc label/index name use karta hai)

but hamare case me index names 0 se start hai isliye hum **df.loc[0:2]** use kar rahe hain 
and isme 2 included rhega (last value included) ✅ **Explicit indexing**

jis number pe wo index hai usko find krne ka tarika hai **df.iloc[0:2]** 
and isme 2 included nhi hoga (last value excluded) ✅ **Implicit indexing**

yaha confusion isliye ho raha hai kyunki index 0 se start hota hai 
aur index name bhi 0,1,2 hi hai, isliye loc aur iloc same lag rhe hain

```python
df.iloc[0:2, ['Name', 'Parch']] #it will throw an error
# ❌ Error: iloc is position-based (only integers allowed for rows & columns),
# but ['Name','Parch'] are column labels (strings), not positions
```

```python
df.loc[0:2, ['Name', 'Parch']]
# ✅ Works: loc is label-based, so it accepts index labels (0:2) and column names
# also includes row 2 because loc slicing is inclusive
```

```python
#imputation of missing value(fill the missing values with some number or string)
#for numeric data you can replace with mean and median, if the outlier treatment done then you have to impute the missing values with mean else median
#categorical data replace with mode
#impute missing values may be with some constant >>0 (or random numbers)


#--pw
```

```python
# 🔥 Missing values (NaN) ko fill karna = Imputation

# 1. Numeric data (numbers)
# → mean (average) ya median (middle value) use karte hain

# rule:
# → agar outliers (extreme values) handle kar liye → use mean
# → agar outliers abhi bhi hain → use median (safe option)

# 2. Categorical data (text like male/female, city, etc.)
# → mode (most frequent value) use karte hain

# 3. Constant value
# → kabhi-kabhi fixed value fill kar dete hain (like 0, "Unknown")

# 4. Random values
# → kabhi distribution ke according random values bhi fill karte hain (advanced case)
```

👉 Numeric (numbers)

- mean = average
- median = safe (jab data messy ho)

👉 Categorical (text)

- mode = sabse common value

⚡ Easy Example

Age column:

[20, 22, NaN, 25]

mean → (20+22+25)/3 = 22.33  
 
median → 22

City column:

['Delhi', 'Mumbai', NaN, 'Delhi']

mode → Delhi

```python
data1 = {
    'A': [1, 2, None, 4, 5, None, 7, 8, 9, 10],
    'B': [None, 11, 12, 13, None, 15, 16, None, 18, 19]
}

df2 = pd.DataFrame(data1)
df2
```

```python
df2.A.fillna(df2['A'].mean())
# fills missing (NaN) values in column 'A' with the mean (average) of that column
# returns a new Series (original column unchanged unless assigned back)

#'A': [1, 2, None, 4, 5, None, 7, 8, 9, 10]
#(mean)A = (1 + 2 + 4 + 5 + 7 + 8 + 9 + 10) / 8 --- result = 5.75

#result all NaN replace by 5.75 in column 'A' (but original df2['A'] remains unchanged unless you assign it back)
```

```python
df2.A.fillna(df2['A'].median())
# fills missing (NaN) values in column 'A' with the median (middle value) of that column
# returns a new Series (original column unchanged unless assigned back)
```
