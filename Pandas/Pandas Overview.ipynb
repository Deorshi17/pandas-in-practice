## Pandas – Brief Introduction

**Pandas is a Python library for data manipulation and analysis.**

It is mainly used to **load, clean, transform, analyze, and summarize data.**

Most datasets in Data Analytics come as:

- CSV files
- Excel files
- SQL tables
- JSON files

Pandas helps you **convert this raw data into structured tables and analyze it easily.**

The main data structure in Pandas is:

**1️⃣Series** → 1D data (like a column)

**2️⃣DataFrame** → 2D data (like Excel table)

example:

````python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Age": [21, 22, 23]
}

df = pd.DataFrame(data)

# Here df is a DataFrame
````

## Most Important Pandas Topics for Data Analytics

Study these in this order.

### 1️⃣ Creating and Loading Data

You must know how to load datasets.

Important functions:

```
pd.read_csv()
pd.read_excel()
pd.read_json()
pd.DataFrame()
```

Example:

```python
df = pd.read_csv("data.csv")
```
### 2️⃣ Data Inspection (Very Important)

First step in every DA project.

Functions to study:

```
df.head()
df.tail()
df.shape
df.columns
df.info()
df.describe()
df.dtypes
```

These help you **understand the dataset structure.**

### 3️⃣ Selecting Data

You must know how to extract rows and columns.

Important:
```
df["column"]
df[["col1","col2"]]

df.loc[]
df.iloc[]
```

Example:
```python
df["Age"]
df.loc[0]
```
### 4️⃣ Filtering Data

Used heavily in real analysis.

Example:

```python
df[df["Age"] > 25]
```

Multiple conditions:

```python
df[(df["Age"] > 25) & (df["Salary"] > 50000)]
```

### 5️⃣ Handling Missing Data

Very common in real datasets.

Functions:
```
df.isnull()
df.notnull()
df.dropna()
df.fillna()
```
Example:
```python
df.fillna(0)
```
### 6️⃣ Data Cleaning

Fix incorrect data.

Important methods:
```
df.rename()
df.drop()
df.replace()
df.astype()
df.duplicated()
df.drop_duplicates()
```
### 7️⃣ Sorting Data
```
df.sort_values()
df.sort_index()
```
Example:
```python
df.sort_values("Age")
```
### 8️⃣ GroupBy (Extremely Important for DA)

Used for **data aggregation and summaries.**

Example:
```python
df.groupby("Department")["Salary"].mean()
```

Common aggregations:
```
sum()
mean()
count()
min()
max()
```
### 9️⃣ Data Aggregation
```
df.value_counts()
df.unique()
df.nunique()
```
### 🔟 Merging and Joining Data

Important when working with **multiple datasets.**

Functions:
```
pd.merge()
pd.concat()
df.join()
```
### 1️⃣1️⃣ Pivot Tables

Used for **data summarization like Excel Pivot Tables**
```
pd.pivot_table()
```
### 1️⃣2️⃣ Basic Visualization (Optional)
```
df.plot()
```

But normally we use:

Matplotlib

Seaborn
