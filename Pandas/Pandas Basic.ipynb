```python
d
# 👉 Left = custom index (names)
# 👉 Right = values
```

```python
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
pd.read_json(url)
# Read JSON data from the API
# Converts JSON response into a pandas DataFrame
```

```python
import requests # Import the requests library (used to make HTTP/API requests)
data = requests.get(url)
# Send a GET request to the given URL (API endpoint)
# This fetches data from the server (GitHub API in this case)
data
# "data" now stores the full HTTP response object
#<Response [200]>  Means request successful
```

```python
# Convert the response data into JSON format (Python object)
# .json() parses the API response into a list/dictionary
df6= data.json()
df6
# Output will look like a list of dictionaries (not a table)
```

```python
len(df6)
```

```python
# Loop through all items in df6 (df6 is a list of JSON objects)
# range(len(df6)) → generates index values from 0 to (length - 1)
for i in range(len(df6)):
# Access each item using index i
# df6[i] → one issue (dictionary)
    print(df6[i]['user'])
    # 'user' key contains details about the user who created the issue
    # df6[i]['user'] → nested dictionary with user info (login, id, avatar, etc.)
```
