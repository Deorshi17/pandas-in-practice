Why whole table is shown?

df[df['Fare'] < df['Fare'].mean()]

👉 Because:

df[...] → always returns rows of the full DataFrame

You only applied condition on Fare

But you didn’t select any specific column

What’s happening

👉 Condition:

df['Fare'] < mean

👉 This gives True/False for rows

👉 Then:

df[ ... ]

👉 Returns all columns for those matching rows

🧠 Easy understanding

👉 “Condition Fare pe laga hai, par output me poori row dikhegi”

```python
df[(df['Sex'] == 'female') & (df['Fare'] > df['Fare'].mean())]
# Select rows where passenger is female AND fare is greater than the average fare
#or, how many females paid more than avg fare
```

🔍 Step-by-step samjho:

1️⃣ df['Sex'] == 'female'

👉 Sirf female passengers select karta hai

2️⃣ df['Fare'] > df['Fare'].mean()

👉 Jin logon ka fare average se zyada hai

3️⃣ & (AND operator)

👉 Dono condition ek saath true honi chahiye

4️⃣ df[...]

👉 Jo rows dono condition satisfy karti hain, wo return hoti hain
👉 Poora table (all columns) dikhega

```python
len(df[(df['Sex'] == 'female') & (df['Fare'] > df['Fare'].mean())])
# Count the number of female passengers who paid more than the average fare
```

Easy breakdown:

df['Sex'] == 'female' → females

df['Fare'] > mean → above average fare

& → dono condition

len() → count

```python
len(df[(df['Sex'] == 'male') & (df['Fare'] > df['Fare'].mean())])
# Count the number of male passengers who paid more than the average fare
```
