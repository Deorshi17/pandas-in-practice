```python
df2 = pd.DataFrame({"a": [1, 2, 3, 4, 5, 6, 7, 8, 9]})
# Create a DataFrame with one column 'a' and values from 1 to 9
```

```python
df2
```

```python
df2.mean()
```

```python
df2.rolling(window = 1).mean() #Provide rolling window calculations.
# df2 → DataFrame
# rolling(window=1) → ek time pe sirf 1 value uthata hai
# mean() → us value ka average nikalta hai
# since 1 hi value hai → average wahi value hota hai
```

```python
df2.rolling(window = 2).mean()
# df2 → DataFrame
# rolling(window=2) → ek time pe 2 values ka group (sliding window) banata hai
# mean() → un 2 values ka average nikalta hai
# first value ke liye pair nahi banta → isliye NaN aata hai
```

Agar data hai:(for window = 2)

[1, 2, 3, 4]

Groups banenge:
- [1, 2]
- [2, 3]
- [3, 4]

🧠 Mean (average):

- [1, 2] → (1+2)/2 = 1.5
- [2, 3] → (2+3)/2 = 2.5
- [3, 4] → (3+4)/2 = 3.5

```python
df3 = df2.rolling(window = 3).mean()
# df2 → original DataFrame
# rolling(window=3) → ek time pe 3 values ka group (sliding window) banata hai
# mean() → un 3 values ka average nikalta hai
# first 2 rows ke liye 3 values nahi milti → isliye NaN aata hai
# result ko df3 me store kar rahe hain
```

- df2 se 3 values ka average nikaal ke
- result ko df3 me store kar diya

```python
df3.dropna(inplace=True)
# Remove rows with NaN values from df3 permanently
```

```python
# Import datetime class # datetime module se datetime class import kar rahe hain
from datetime import datetime

# date string (text format)
date = "2024-03-29"

# Convert string into actual datetime object
date = datetime.strptime(date, "%Y-%m-%d")
# strptime ka use karke string ko datetime object me convert kar rahe hain
# "%Y-%m-%d" format batata hai: Year-Month-Day (2024-03-29)
```
